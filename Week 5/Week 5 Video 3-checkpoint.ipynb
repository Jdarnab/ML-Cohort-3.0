```python
#train a logistic regression classifier to predict whether a flower is iris virginica or not 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
iris= datasets.load_iris ()
x= iris['data'][:,3:]
y= (iris['target']==2).astype(np.int64)
#train a logisitic regression classifier 
clf= LogisticRegression ()
clf.fit(x,y)
Prediction = clf.predict ([[2.6]])
print (Prediction )
#using matplotlib to plot the visualization taking petal width into consideration 
X= np.linspace (0,3,1000).reshape(-1,1)
Y= clf.predict_proba (X)
plt.plot (X,Y[:,1], 'g',label =' Virginica')
plt.legend(loc='upper right')
plt.show ()
```
